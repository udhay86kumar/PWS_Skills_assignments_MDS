### Anomaly Detection Assignment

# Question 1: What is anomaly detection and what is its purpose?

Anomaly detection is the process of identifying data points that deviate significantly from the majority of the dataset.
Its purpose is to detect rare events, unusual patterns, or data points that do not conform to expected behavior.
Common applications include fraud detection, network security, and fault detection in industrial systems.


# Question 2: What are the key challenges in anomaly detection?

1. **Lack of Labeled Data**: Many real-world datasets lack labeled anomalies.
2. **Data Imbalance**: Anomalies are rare compared to normal instances.
3. **High Dimensionality**: Complex data structures make it difficult to detect anomalies.
4. **Concept Drift**: The definition of normal behavior may change over time.
5. **Noise Sensitivity**: Anomaly detection models can be sensitive to outliers and noise.


# Question 3: How does unsupervised anomaly detection differ from supervised anomaly detection?

- **Supervised Anomaly Detection**: Requires labeled data (normal vs. anomaly) and uses classification models.
- **Unsupervised Anomaly Detection**: Does not require labeled data and detects anomalies based on deviations from normal patterns.


# Question 4: What are the main categories of anomaly detection algorithms?

1. **Statistical Methods**: Gaussian distribution, z-score, etc.
2. **Distance-Based Methods**: k-Nearest Neighbors (KNN), Local Outlier Factor (LOF).
3. **Density-Based Methods**: DBSCAN, LOF.
4. **Machine Learning Models**: Isolation Forest, Autoencoders, One-Class SVM.
5. **Spectral Methods**: PCA-based anomaly detection.


# Question 5: What are the main assumptions made by distance-based anomaly detection methods?

1. Anomalies are located in low-density regions of the feature space.
2. Normal points have many close neighbors, while anomalies have few neighbors.
3. Distance metrics (e.g., Euclidean, Manhattan) accurately reflect similarity.


# Question 6: How does the LOF algorithm compute anomaly scores?

LOF (Local Outlier Factor) computes anomaly scores by:
1. Measuring the local density of a data point based on k-nearest neighbors.
2. Comparing its density with the density of its neighbors.
3. If a point has significantly lower density than its neighbors, it is an anomaly.

# Question 9: Using the Isolation Forest algorithm with 100 trees and a dataset of 3000 data points, what is the anomaly score for a data point that has an average path length of 5.0 compared to the average path length of the trees?

The anomaly score is computed as:

s(x, n) = 2^(-E(h(x))/c(n))

where:
- E(h(x)) is the average path length (5.0 in this case).
- c(n) is the average path length of the trees for n=3000 data points, approximately c(3000) ≈ 9.95.

s(x, 3000) = 2^(-5.0/9.95) ≈ 0.74

Since the score is close to 1, the data point is likely to be an anomaly.
