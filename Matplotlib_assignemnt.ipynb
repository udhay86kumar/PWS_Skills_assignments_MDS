## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Matplotlib is a popular and versatile plotting library in Python used for creating static, interactive, and animated visualizations in a variety of formats and styles. It provides a high-level interface for creating a wide range of plots and charts, making it an essential tool for data visualization in various fields, such as data analysis, scientific research, engineering, and more.

Matplotlib is used for:

Data Visualization: Matplotlib allows you to create various types of plots to visually represent data, patterns, and relationships within your datasets.

Exploratory Data Analysis: It helps in exploring the characteristics of your data and identifying trends, outliers, and correlations.

Presentation and Communication: Matplotlib plots are highly customizable, making it suitable for creating publication-quality graphics for presentations, reports, and articles.

Model Evaluation: Visualization of model outputs, such as confusion matrices, ROC curves, and learning curves, aids in evaluating and understanding the performance of machine learning models.

Educational Purposes: Matplotlib is also used in educational contexts to teach concepts related to data visualization and programming.

The pyplot module of Matplotlib is a sub-library that provides a simple and procedural way to create and customize plots. Here are five types of plots that can be plotted using the pyplot module of Matplotlib:

Line Plot: A basic plot showing the relationship between two variables with points connected by lines. Useful for visualizing trends over time or continuous data.

Bar Plot: Used to compare categories or groups. It displays rectangular bars whose lengths are proportional to the values they represent.

Scatter Plot: Displays individual data points as markers on a 2D plane. Useful for exploring the relationship between two continuous variables.

Histogram: Represents the distribution of a dataset by dividing it into intervals (bins) and counting the number of observations in each bin.

Pie Chart: A circular chart divided into sectors, where each sector represents a proportion of the whole. Often used to show the composition of a categorical variable.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("scatter_plot")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It's particularly useful when you want to display multiple plots side by side for comparison or visualization purposes. Each subplot can display a different plot type or data.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Line Plot 1')

plt.subplot(2,2,2)
plt.plot(x1,y1)
plt.title('Line Plot 2')

plt.subplot(2,2,3)
plt.plot(x2,y2)
plt.title('Line Plot 3')

plt.subplot(2,2,4)
plt.plot(x3,y3)
plt.title('Line Plot 2')

plt.tight_layout()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of data visualization that uses rectangular bars to represent categorical data. Each bar's length or height is proportional to the value it represents, allowing for easy comparison between different categories. Bar plots are used to display and compare the frequency, count, or magnitude of different categories in a dataset.

Bar plots are particularly useful in the following scenarios:

Comparisons: Bar plots allow you to visually compare the values of different categories easily. For example, you can use a bar plot to compare sales of different products, performance of different teams, or frequency of different events.

Distribution: Bar plots can help you visualize the distribution of categorical data. For instance, you can use a bar plot to show the distribution of grades in a class or the distribution of customer ratings.

Ranking: Bar plots can also show the ranking of different categories based on their values. This is common in scenarios where you want to show which category is the most or least popular.

Composition: Bar plots are suitable for displaying the composition of a whole, where each bar represents a part of the whole. For example, a bar plot can show the percentage of different types of expenses in a budget.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot')
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
A box plot, also known as a box-and-whisker plot, is a graphical representation used to display the distribution and summary statistics of a dataset. It provides a compact way to visualize the central tendency, spread, and potential outliers within the data. A box plot is especially useful for comparing distributions across different categories or groups.

A typical box plot consists of several components:

Box: The central rectangle of the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The bottom edge of the box represents the first quartile (Q1), and the top edge represents the third quartile (Q3).

Whiskers: Lines extending from the edges of the box, called whiskers, represent the range of the data within a certain threshold. Typically, the whiskers extend to the minimum and maximum data points within 1.5 times the IQR from the quartiles. Data points beyond this threshold are considered potential outliers.

Outliers: Data points that fall outside the whiskers are plotted as individual points. These points are often considered potential outliers, indicating values that are significantly different from the rest of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
```

```python
plt.boxplot(data)
plt.xlabel("Boxs")
plt.ylabel("Numbers")
plt.xticks([1,2],['box1','box2'])
```
